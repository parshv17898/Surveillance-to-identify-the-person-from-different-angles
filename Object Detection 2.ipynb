```python
predicted1 = np.zeros((746,2))
predicted2 = np.zeros((743,2))
```

```python
 def generate(image_path,video,num):
        face_cascade = cv2.CascadeClassifier('/content/drive/My Drive/Sem 7/Minior/meet parshv/cascade.xml')
        img1 = Image.open(image_path)
        img=np.array(img1)
        myimg = cv2.imread(image_path)
        if (img is None):
            print("Can't open image file")
            return 0

        #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(img, 1.1, 3, minSize=(100, 100))
        if (len(faces)==0):
            print('Failed to detect face')
            myimg=cv2.resize(myimg,(1024,700), interpolation = cv2.INTER_AREA)
            return myimg

            cv2.destroyAllWindows()

        facecnt = len(faces)
        print("Detected faces: %d" % facecnt)
        i = 0
        height, width = img.shape[:2]
        lastimg = 0
        for (x, y, w, h) in faces:
            r = max(w, h) / 2
            centerx = x + w / 2
            centery = y + h / 2
            nx = int(centerx - r)
            ny = int(centery - r)
            nr = int(r * 2)
            faceimg = img[ny:ny+nr, nx:nx+nr]
            lastimg = cv2.resize(faceimg, (32, 32))

            a=np.array(lastimg)
            a = np.expand_dims(a, axis=0)
            prediction = model.predict(a)    
            print(prediction)
            if prediction[0][0]>0.5:
                if video==1:
                    predicted1[num][0]=round((prediction[0][0])*100, 2)
                    print(predicted1[num][0])
                if video==2:
                    predicted2[num][0]=round((prediction[0][0])*100, 2)
                    print(predicted2[num][0])
            elif prediction[0][0]<0.5:
                if video==1:
                    predicted1[num][1]=round((prediction[0][1])*100, 2)
                    print(predicted1[num][1])
                if video==2:
                    predicted2[num][1]=round((prediction[0][1])*100, 2)
                    print(predicted2[num][1])


                                    
            i = i + 1

        return i
```

```python
for i in range(746):
    generate("/content/drive/My Drive/Sem 7/Minior/meet parshv/Adv/data1/frame"+str(i)+".jpg",1,i)
    
```

```python
from os.path import isfile, join

def convert_frames_to_video(pathIn,pathOut,fps):
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]

    #for sorting the file names properly
    files.sort(key = lambda x: int(x[5:-4]))

    for i in range(729):
        filename=pathIn + files[i]
        #reading each files
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        print(filename)
        #inserting the frames into an image array
        frame_array.append(img)

    out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)

    for i in range(len(frame_array)):
        # writing to a image array
        out.write(frame_array[i])
    out.release()
```

```python
#Converting Frames of Predicted_data to Predicted_Video.mp4
def main():
    pathIn= '/content/drive/My Drive/Sem 7/Minior/meet parshv/Adv/predicted_data/'
    pathOut = '/content/drive/My Drive/Sem 7/Minior/meet parshv/Adv/predicted_video.mp4'
    fps = 15.0
    convert_frames_to_video(pathIn, pathOut, fps)

if __name__=="__main__":
    main()
```
