```python
 def generate(image_path):
        face_cascade = cv2.CascadeClassifier('/content/drive/My Drive/Sem 7/Minior/meet parshv/cascade.xml')
        img1 = Image.open(image_path)
        img=np.array(img1)
        myimg = cv2.imread(image_path)
        if (img is None):
            print("Can't open image file")
            return 0

        #img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(img, 1.1, 3, minSize=(100, 100))
        if (len(faces)==0):
            print('Failed to detect face')
            myimg=cv2.resize(myimg,(1024,700), interpolation = cv2.INTER_AREA)
            return myimg

            cv2.destroyAllWindows()

        facecnt = len(faces)
        print("Detected faces: %d" % facecnt)
        i = 0
        height, width = img.shape[:2]
        lastimg = 0
        for (x, y, w, h) in faces:
            r = max(w, h) / 2
            centerx = x + w / 2
            centery = y + h / 2
            nx = int(centerx - r)
            ny = int(centery - r)
            nr = int(r * 2)

            faceimg = img[ny:ny+nr, nx:nx+nr]
            lastimg = cv2.resize(faceimg, (32, 32))

            a=np.array(lastimg)
            a = np.expand_dims(a, axis=0)
            prediction = model.predict(a)    
            print(prediction)
            
            if prediction[0][0]>0.5:
                #If Meet is predicted, mark rectangle and put text label above it.
                faceimg=cv2.rectangle(myimg,(x,y),(x+w,y+h),(0,255,0),5)
                faceimg=cv2.putText(faceimg,'Meet-'+str(round((prediction[0][0])*100, 2)), (x,y), cv2.LINE_AA, 1.0, (255, 0, 0),3)
                faceimg=cv2.resize(myimg,(1024,700), interpolation = cv2.INTER_AREA)
            elif prediction[0][0]<0.5:
                #If Parshv is predicted, mark rectangle and put text label above it.
                faceimg=cv2.rectangle(myimg,(x,y),(x+w,y+h),(0,0,255),5)
                faceimg=cv2.putText(faceimg,'Parshv-'+str(round((prediction[0][1])*100, 2)), (x,y), cv2.LINE_AA, 1.0, (0, 0, 255),3)
                faceimg=cv2.resize(myimg,(1024,700), interpolation = cv2.INTER_AREA)
                                    
            i = i + 1

        return faceimg
```

```python
#Converting Frames to Video
from os.path import isfile, join

def convert_frames_to_video(pathIn,pathOut,fps):
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
    
    
    #for sorting the file names properly
    files.sort(key = lambda x: int(x[5:-4]))
    
    for i in range(len(files)):
        filename=pathIn+'/'+files[i]
       
        
        #reading each files
        img = cv2.imread(filename)
     #   print(img)
        height, width, layers = img.shape
        size = (width,height)
        
      #  print(filename)
        #inserting the frames into an image array
        frame_array.append(img)

    out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)

    for i in range(len(frame_array)):
        # writing to a image array
        out.write(frame_array[i])
    out.release()

```

```python
#Converting Frames of Predicted_data1 to Predicted_Video1.mp4
def main():
    pathIn= '/content/drive/My Drive/Sem 7/Minior/meet parshv/Adv/predicted_data1'
    pathOut = '/content/drive/My Drive/Sem 7/Minior/meet parshv/Adv/Predicted_video1.mp4'
    fps = 15.0
    convert_frames_to_video(pathIn, pathOut, fps)

if __name__=="__main__":
    main()
```
